##Indexing and Slicing

```python
l1=[1,2,3,"data",[10,20],[8,0,5]]
```

```python
# extract "data" from l1
l1[3]
```

```python
# listname[startnum:stopnum : step]
# extract 2, 3,"data"
l1[1:4]
```

```python
# extract 20 from l1
l1[4][1]
```

```python
# extract 0,5 from l1
l1[5][1:]
```

```python
# extract 20 from l1 in negative
l1[-2][-1]
```

```python
# extract 0,5 from l1 using negative slicing
l1[-1][-2:]
```

```python
# we can also use both negative & positive slicing
l1[-1][1:]
```

```python
# reverse
l1[::-1]
```

```python
l1
```

```python
l1[-1][::-1]
```

```python
l1
```
