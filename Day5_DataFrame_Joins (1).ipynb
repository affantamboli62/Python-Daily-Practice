#Merge
- Merge is used to combine Dataframes based on columns with columns(similar to join operations)
- outer join : use union of keys
- inner join : interaction of keys
- left join : use keys from left table/dataframe/object
- right join : use keys from right object

```python
import pandas as pd
df1=pd.DataFrame({"city":["new york","chicago"],
                  "temperature":[21,14]})
df1
```

```python
df2=pd.DataFrame({"city":["new york","orlando","chicago"],
                  "humidity":[21,75,23],
                  "temperature":[21,75,19]})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df4=pd.merge(df1,df2,on="temperature",how="inner")
df4
```

```python
df5=pd.merge(df1,df2,on="city",how="outer")
df5
```

####indicator =True

```python
df5=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df5
```

```python
d1=pd.DataFrame({"emp_id":[1,2,3,4],
                 "emp_name":["Affan","Tanmay","Vijay","Rohit"]})
d1
```

```python
d2=pd.DataFrame({"emp_id":[1,6,3,5],
                 "salary":[1255,475,8552,852]})
d2
```

```python
d3.loc[d3["emp_name"]=="Affan"]
```

```python
d4=pd.merge(d1,d2,on="emp_id",how="outer")
d4
```

```python
df1=pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature":[32,45,30,45],
    "humidity":[80,60,78,65]
})
df1
```

```python
df2=pd.DataFrame({
    "city":["chicago","new york","sandiego"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
df2
```

#Suffixes

```python
df3=pd.merge(df1,df2,on="city",how="outer",suffixes=("_first","_second"))
df3
```
