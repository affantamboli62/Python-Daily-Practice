####Operators
- union()/| : Multiple set combine in one set
- intersection() / & : common elements from sets
- difference() / - : elements in one set but not others
- symmetric_difference() / EX-OR(^) : uncomman elements shown in set

```python
s1={2,3,1,2}
s2={6,5,3,21,1}
s3={6,2,9,10,20}
```

```python
s1 | s2
```
